```python
# Importing all the essentails Libraies 
import pandas as pd
import requests
#Data Collection
url = "https://en.wikipedia.org/wiki/COVID-19_pandemic_by_country_and_territory"
req = requests.get(url)
```

```python
# using read_Html() of pandas library to look for all the tablular data on the web page 
# data_list has all the list of dataframes
data_list = pd.read_html(req.text)
##print(data_list)
# Lets extract the required datafram from data_list
#target_df = data_list[1]
#print(target_df)
# Lets extract the required datafram from data_list
target_df = data_list[12]
#print(target_df)

```

```python
# Data Cleaning 
target_df.head()


```

```python
# Issue 1 Columns Names
# renaming colimns
target_df.columns = ['col0','Country Name','Total Deaths/Million','Total Deaths','Total Cases']
#target_df.head()

#Issue 2 
# Getting rid of extra columns like col0
target_df = target_df[['Country Name','Total Deaths/Million','Total Deaths','Total Cases']]
target_df.head()   


```

```python
#Issue 3 Getting rid of last  rows as they don't have much usefull data. 
last_index = target_df.index[-1]
target_df = target_df.drop([last_index,last_index-1])
target_df





```

```python
target_df[1:]
```

```python
#Issue 4 
#Removing Inconsistent Country Names
```

```python
#Issue 4 
#Removing Inconsistent Country Names

target_df['Country Name'] = target_df['Country Name'].str.replace('\[.*\]','')
```

```python
target_df['Country Name'] 
```

```python
# Issue 5
#Replacing 'No Data'or '-', with '0' ; Taking both parametes as sting type
```

```python
target_df['Total Deaths/Million'] = target_df['Total Deaths/Million'].str.replace('—','0')
target_df['Total Deaths'] = target_df['Total Deaths'].str.replace('—','0')
target_df['Total Cases'] = target_df['Total Cases'].str.replace('—','0')
target_df
```

```python
#Issue 6
# Fixing the wrong dtypes 
target_df.dtypes
```

```python
#Issue 6
# Fixing the wrong dtypes
target_df['Total Cases'] = pd.to_numeric(target_df['Total Cases'])
target_df['Total Deaths/Million'] = pd.to_numeric(target_df['Total Deaths/Million'])
target_df['Total Deaths'] = pd.to_numeric(target_df['Total Deaths'])
target_df.dtypes
```

```python
#Data organization - export data in excel or CSV
target_df.to_csv(r'Covid19_dataset1.csv',index=False,header = True)
#target_df.to_excel(r'Covid19_dataset.xlsx',index=False,header = True, encoding='utf-8')
```
